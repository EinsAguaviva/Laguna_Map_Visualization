#### GADM (Database of Global Administrative Areas):

GADM is a spatial database of the location and shape of administrative areas, such as countries, states, provinces, and municipalities, for all countries of the world. It provides detailed administrative boundaries at various levels, making it a valuable resource for geographic information systems (GIS) applications, spatial analysis, and research purposes. GADM is widely used by researchers, policymakers, and developers for tasks such as mapping, spatial analysis, and demographic studies. You can access the data [here on their official website.](https://gadm.org/index.html)

#### GeoPandas:

GeoPandas is an open-source Python library that extends the capabilities of the Pandas library to allow for easy manipulation and analysis of geospatial data. It provides data structures and functions to work with geospatial data seamlessly within the familiar Pandas framework. Read more about them [here!](https://geopandas.org/en/stable/)

#### Our First Map Visualization:

Let's get the map of the Philippines as an example:
<br>
We can download the GADM data for the Philippine shape file [here.](https://gadm.org/download_country.html)

We can see different GeoJSON levels from 0-3. Below are their breakdowns.
<br>
- **Level 0**: This typically represents the entire country as a single polygon. It's the highest level of administrative division, encompassing the entire territory of the Philippines.

- **Level 1**: This represents the first-level administrative divisions, which in the case of the Philippines are the regions. The Philippines is divided into several regions, each consisting of multiple provinces.

- **Level 2**: This represents the second-level administrative divisions, which are the provinces. Each region in the Philippines is divided into provinces.

- **Level 3**: This represents the third-level administrative divisions, which are the municipalities or cities (depending on the classification). Provinces in the Philippines are further divided into municipalities and cities.

```python
# this is how we import geopandas
import geopandas as gpd
```

We then read the geojson file we downloaded from GADM.

```python
gdf_0 = gpd.read_file("./Data/level_0/gadm41_PHL_0.json")
```

```python
display(gdf_0)
```

| Column    | Description                               |
|-----------|-------------------------------------------|
| GID_0     | Unique identifier for the country at Level 0 |
| COUNTRY   | Name of the country (Philippines)            |
| geometry  | Geometric data describing the country's shape  |

Here is the data dictionary of the level 0 gadm file that we read from json.

```python
# Basic plot, single color
gdf_0.plot();
```

You can visit the plot function documentation on this [link](https://geopandas.org/en/stable/docs/reference/api/geopandas.GeoDataFrame.plot.html#geopandas.GeoDataFrame.plot)

```python
# We can increase the figsize and use plt.show() like matplotlib
gdf_0.plot(figsize=(15,15));
```

```python
# We can also only get the boundary of the shape file by using the .boundary function
gdf_0.boundary.plot(figsize=(15,15,));

# plt.savefig('ph_boundary.png', dpi=300, bbox_inches='tight')
plt.xticks([])
plt.yticks([])

# Remove the border
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

# Set transparent background
plt.gcf().set_facecolor('none')

# Save the plot to a file
plt.savefig('ph_more_beautiful.png', dpi=300, bbox_inches='tight', transparent=True)
```

```python
# We can also change the color and remove the ticks for more aesthetic visualization
# Notice that we used matplotlib's functions in removing the x and y ticks
gdf_0.boundary.plot(figsize=(15,15,), color='green')

plt.xticks([])
plt.yticks([])

# Remove the border
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

# Set transparent background
plt.gcf().set_facecolor('none')

plt.title('Map of the Philippines');
```
