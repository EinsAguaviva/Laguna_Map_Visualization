***

```python
data = pd.read_csv('accident data.csv')
```

<a name="dat_prep"></a>
## Data Preprocessing
***

```python
fl.table_caption("Road Accident Data",
                 "Displaying the first 5 rows of the Road Accident Data.")

display(data.head())
```

```python
# Dropping duplicated rows and missing values
data.isnull().sum()
data.drop_duplicates(inplace=True)
data.dropna(inplace=True)
```

```python
# Rename columns
data.rename(columns={'Accident Date': 'Accident_Date', 'District Area': 'District_Area'}, inplace=True)

# Replace values in columns
data.replace({'Vehicle_Type': {'Car': 'Car',
                               'Van / Goods 3.5 tonnes mgw or under': 'Van',
                               'Bus or coach (17 or more pass seats)': 'Bus',
                               'Motorcycle over 500cc': 'Motorcycle (>500cc)',
                               'Goods 7.5 tonnes mgw and over': 'Over 7.5 Tonnes',
                               'Motorcycle 125cc and under': 'Motorcycle (<125cc)',
                               'Taxi/Private hire car': 'Taxi',
                               'Motorcycle over 125cc and up to 500cc': 'Motorcycle (125-500cc)',
                               'Motorcycle 50cc and under': 'Motorcycle (<50cc)',
                               'Goods over 3.5t. and under 7.5t': 'Tonnes (3.5-7.5t)',
                               'Other vehicle': 'Other',
                               'Minibus (8 - 16 passenger seats)': 'Minibus',
                               'Agricultural Vehicle': 'Agricultural Vehicle',
                               'Pedal cycle': 'Bike',
                               'Data missing or out of range': 'None',
                               'Ridden horse': 'Horse'},
              'Road_Surface_Conditions': {'Dry': 'Dry', 'Wet or damp': 'Wet', 'Snow': 'Snow', 'Frost or ice': 'Ice', 'Flood over 3cm. deep': 'Flood'},
              'Weather_Conditions': {'Fine no high winds': 'Fine (NHW)', 'Raining no high winds': 'Raining (NHW)', 'Other': 'Other', 'Fine + high winds': 'Fine (HW)', 'Raining + high winds': 'Raining (HW)', 'Snowing no high winds': 'Snowing (NHW)', 'Fog or mist': 'Fog/Mist', 'Snowing + high winds': 'Snowing (HW)'}
             }, inplace=True)
```

```python
# Dropping 'Other' and 'None'
data = data[(data['Vehicle_Type'] != 'Other') & (data['Weather_Conditions'] != 'Other') & (data['Vehicle_Type'] != 'None')]
```

```python
fl.table_caption("Revised Road Accident Data",
                 "Displaying the changes of the first 5 rows of the Road Accident Data.")

data[['Accident_Severity', 'Latitude', 'Longitude', 'Number_of_Casualties', 'Number_of_Vehicles', 'Road_Surface_Conditions', 'Weather_Conditions', 'Vehicle_Type']].head()
```

```python
# Creating a uniform color palette:
severity_colors = {'Fatal': '#a70000', 'Serious': '#ff9a00', 'Slight': '#607c3c'}
```

<a name="dat_expl"></a>
## Data Exploration
***

### Number of Casualties and Vehicles

```python
plt.figure(figsize=(8, 6))
plt.plot(data.groupby('Number_of_Vehicles')['Number_of_Casualties'].sum(), marker='o', linestyle='-', markeredgecolor='black')
plt.xlabel('Number of Vehicles')
plt.ylabel('Total Number of Casualties')
plt.title('Total Casualties per Number of Vehicles')
plt.grid(True)
plt.show()

fl.fig_caption("Number of Casualties per Number of Vehicle",
               "Line chart of the total casualties per number of vehicles.")
```

### Accident Severity and Road Surface Conditions

```python
count_data = data.groupby(['Road_Surface_Conditions', 'Accident_Severity']).size().reset_index(name='Count')

plt.figure(figsize=(10, 7))
sns.lineplot(data=count_data.pivot_table(index='Road_Surface_Conditions', columns='Accident_Severity', values='Count', fill_value=0), marker='o', hue_order=['Fatal', 'Serious', 'Slight'], palette=severity_colors)
plt.yticks(range(0, 510000, 50000), [f'{i:,}' for i in range(0, 510000, 50000)])
plt.title('Relationship between Accident Severity and Road Surface Conditions')
plt.xlabel('Road Surface Conditions')
plt.ylabel('Number of Casualties')
plt.legend(title='Accident Severity')
plt.show()

fl.fig_caption("Accident Severity in Road Surface Conditions",
               "Line plot depicting the correlation between Accident Severity and Road Surface Conditions.")
```

### Accident Severity and Weather Conditions

```python
count_data_weather = data.groupby(['Weather_Conditions', 'Accident_Severity']).size().reset_index(name='Count')

plt.figure(figsize=(10, 7))
sns.lineplot(data=count_data_weather.pivot_table(index='Weather_Conditions', columns='Accident_Severity', values='Count', fill_value=0), marker='o', hue_order=['Fatal', 'Serious', 'Slight'], palette=severity_colors)
plt.yticks(range(0, 510000, 50000), [f'{i:,}' for i in range(0, 510000, 50000)])
plt.title('Relationship between Accident Severity and Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Casualties')
plt.legend(title='Accident Severity')
plt.xticks(rotation=45)
plt.show()

fl.fig_caption("Accident Severity in Weather Conditions",
               "Line plot illustrating the correlation between Accident Severity and Weather Conditions, with annotations for High Winds (HW) and No High Winds (NHW).")
```

### Accident Severity and Vehicly Type

```python
count_data_vehicle = data.groupby(['Vehicle_Type', 'Accident_Severity']).size().reset_index(name='Count')

plt.figure(figsize=(11, 7))
sns.lineplot(data=count_data_vehicle.pivot_table(index='Vehicle_Type', columns='Accident_Severity', values='Count', fill_value=0), marker='o', hue_order=['Fatal', 'Serious', 'Slight'], palette=severity_colors)
plt.yticks(range(0, 510000, 50000), [f'{i:,}' for i in range(0, 510000, 50000)])
plt.title('Relationship between Accident Severity and Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Number of Casualties')
plt.legend(title='Accident Severity')
plt.xticks(rotation=60)
plt.show()

fl.fig_caption("Accident Severity in Vehicle Types",
               "Line plot illustrating the relationship between Accident Severity and Vehicle Type.")
```

### Casualties and Vehicles by Accident Severity

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Number_of_Vehicles', y='Number_of_Casualties', hue='Accident_Severity', data=data, hue_order=['Fatal', 'Serious', 'Slight'], palette=severity_colors, alpha=0.5)
plt.title('Number of Casualties and Vehicles by Accident Severity')
plt.show()

fl.fig_caption("Casualties and Vehicles by Accident Severity",
               "Scatterplot of the correlation between the number of vehicle types and the severity of road accidents.")
```

### Geographical Distribution of Accidents

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Longitude', y='Latitude', hue='Accident_Severity', data=data, hue_order=['Fatal', 'Serious', 'Slight'], palette=severity_colors, alpha=0.5)
plt.title('Geospatial Distribution of Accidents by Severity')
plt.show()

fl.fig_caption("Geographical Distribution of Accidents",
               "Geospatial Distribution of Accidents in District Areas in London.")
```

## Results and Discussions
***

The analysis of factors influencing road accident severity sheds light on the complex dynamics involving various aspects such as the number of vehicles involved, road surface conditions, weather conditions, vehicle types, and geographical distribution. These insights are important for understanding the factors contributing to accident severity and devising effective strategies to mitigate casualties and improve road safety.

*<b>Number of Vehicles:* Accidents involving fewer vehicles tend to result in higher casualties, suggesting a disproportionate impact on casualties, possibly due to higher speeds compared to congested road conditions.

*<b>Road Surface Conditions:* Dry roads are most associated with accidents, with casualties primarily resulting from slight accidents, while wet or damp roads follow a similar pattern. This indicates that casualties are predominantly of slight severity during both dry and wet road conditions.

*<b>Weather Conditions:* Fine weather with no high winds shows the highest number of casualties, primarily with slight severity, followed by rainy weather with no high winds. Casualties during fine weather are predominantly slight, while rainy weather also exhibits evident casualties but with slight severity.

*<b>Vehicle Types:* Cars are identified as the vehicle type with the highest count of casualties, primarily with slight severity. Vans follow a similar pattern. Overall, cars account for the highest number of casualties, mainly with slight severity.

*<b>Correlation between Vehicles and Severity:* Accidents involving two vehicles result in increased casualties, mainly with slight severity, while a higher number of vehicles involved leads to fewer casualties, with almost no slight severity observed.

*<b>Geographical Distribution:* The majority of district areas in London experience almost slight accident severity, indicating a consistent trend of low or slight accident severity across London districts.

<a name="conc"></a>
## Conclusion and Recommendations
***

In conclusion, the analysis highlights key factors influencing road accident severity, including the *number of vehicles involved, road surface* and *weather conditions, and vehicle types*. Accidents with fewer vehicles tend to result in higher casualties, possibly due to higher speeds. <b>Fine weather with no high winds </b> and rainy weather without high winds are associated with the highest number of casualties, primarily with slight severity. <b>Dry and wet road conditions</b> contribute to the highest number of casualties, mostly slight. <b>Cars</b> account for the highest casualties, primarily with slight severity. Accidents involving <b>two vehicles</b> have increased casualties, mainly slightly, while more vehicles result in fewer casualties. Most London districts experience <b>slight accident severity</b>, suggesting a need for preventive measures. Recommendations include speed reduction on less congested roads, improved safety protocols in various weather conditions, promotion of safe driving practices, and infrastructure enhancements across London.
