```python
# Univariate analysis for Gender
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")
sns.countplot(x='gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
print("Gender Distribution:")
print(gender_counts)
print("\n")
fl.fig_caption("Gender Distribtuion Bar Graph",
                "Bar graph showing gender distribution in Barangay Vizwal")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Printed summary
print("Age Distribution Summary:")
print(df['age'].describe())
print("\n")

fl.fig_caption("Age Distribution Histogram",
                "Histogram showing age distribution in Barangay Vizwal")
```

```python
hypertension_counts = df['hypertension'].value_counts()
plt.figure(figsize=(8, 5))

plt.pie(hypertension_counts, labels=hypertension_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
plt.title('Hypertension Distribution')
plt.show()

print("Hypertension Distribution:")
print(hypertension_counts)
print("\n")

fl.fig_caption("Hypertension Distribution Pie Chart",
                "Pie chart showing hypertension distribution in Barangay Vizwal")

#zero has no hypertension and 1 has hypertension
```

```python
hypertension_counts = df['hypertension'].value_counts()
plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")
sns.countplot(x='hypertension', data=df, palette='muted')
plt.title('Hypertension Distribution')
plt.xlabel('Hypertension (0: No, 1: Yes)')
plt.ylabel('Count')
plt.show()
print("Hypertension Distribution:")
print(hypertension_counts)
print("\n")

fl.fig_caption("Hypertension Distribution Bar Graph",
                "Bar graph showing hypertension distribution in Barangay Vizwal")
```

```python
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.countplot(x='smoking_status', data=df, palette='muted')
plt.title('Smoking Status Distribution')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.show()
print("Smoking Status Distribution:")
print(df['smoking_status'].value_counts())
print("\n")

fl.fig_caption("Smoking Status Distribution Bar Graph",
                "Bar graph showing smoking status distribution in Barangay Vizwa")
```

```python
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.histplot(df['bmi'], bins=20, kde=True, color='salmon')
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
print("BMI Distribution Summary:")
print(df['bmi'].describe())
print("\n")

fl.fig_caption("BMI Distribution Histogram",
                "HIstogram showing BMI distribution in Barangay Vizwa")
```
